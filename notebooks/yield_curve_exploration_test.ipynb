## Yield Data

```python
# We won't be using a list, we will have to extract and put into a different format.

maturities = [0.25, 0.50, 1.00, 2.00, 5.00, 10.00]
zero_rates = [0.042, 0.043, 0.045, 0.052, 0.055, 0.060]

curve = YieldCurve (
    maturities = maturities, 
    zero_rates = zero_rates,
    compounding = "continuous"
)
```

```python
test_maturities = [0.75, 3.00, 7.00]

for T in test_maturities:
    z = curve.get_zero_rate(T)
    d = curve.get_discount_factor(T)
    print (f"T = {T:.2f} years | zero rate = {z:.2%} | discount factor = {d:.4%}")
```
